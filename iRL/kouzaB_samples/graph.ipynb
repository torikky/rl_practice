# matpoltlib でグラフを描く

# シンプルなグラフ

```python
# list B-1-(1)
%matplotlib inline
import matplotlib.pyplot as plt  # matplotlib.pyplot のインポート (A)
import numpy as np  # NumPyのインポート (B)
```

```python
# list B-1-(2)

x = np.array([1, 2, 3, 4])
y = np.array([10, 11, 10, 12])
plt.plot(x, y)  # グラフを登録 (A)
plt.show()      # グラフを表示 (B)
```

# 関数のグラフを描く

```python
# list B-1-(3)

def myfunc(x):
    y = (x+3) * x * (x-2)
    return y

# 動作テスト
print(myfunc(1))
```

```python
# list B-1-(4)

x = np.array([0, 1, 2])  # xをベクトルで定義
y = myfunc(x)            # ベクトルも入力にできる
print(y)                 # 出力もベクトルになる
```

```python
# list B-1-(5)

x = np.linspace(-3, 3, 7) # -3から3までの7点を作成 
print(x)
```

```python
# list B-1-(6)

y = myfunc(x)
print(y)
```

```python
# list B-1-(7)

plt.plot(x, y)
plt.show()
```

# 線の種類を変える

```python
# list B-1-(8)

plt.plot(x, y, marker='*', linestyle=':', color='r',
         linewidth=3, markersize=15)
plt.show()
```

```python
# list B-1-(9)

plt.plot(x, y, marker='*', ls=':', c='r', lw=3, ms=15)
plt.show()
```

```python
# list B-1-(10)

plt.plot(x, y, '*:r', lw=3, ms=15)
plt.show()
```

# グラフをカスタマイズする

```python
# list B-1-(11)

x = np.linspace(-3, 3, 50)      # 50個のx点を作成
y = myfunc(x)                   # yを作成
y2 = -6 * x                     # y2を作成

plt.figure(figsize=(6, 3))      # 図のサイズを指定 (A)
plt.plot(x, y, '-k', lw=2, label='original')  # labelの指定(B)
plt.plot(x, y2, ':r', lw=2, label='slope')  # labelの指定(C)
plt.ylim(-10, 20)               # 縦軸の表示範囲 (D)
plt.xlim(-3, 3)                 # 横軸の表示範囲 (E)
plt.yticks([-10, 0, 10, 20])    # 縦軸の目盛り (F)
plt.xticks(range(-3, 4, 1))     # 横軸の目盛りをrange()で指定 (G)
plt.title('$y=(x-3)x(x+2)=x^3-x^2-6x$')  # タイトル表示 (H)
plt.xlabel('x')                 # 横軸のラベル表示 (I)
plt.ylabel('y')                 # 縦軸のラベル表示 (J)
plt.grid(axis='both')           # グリッド表示 (K)
plt.legend(loc='lower right')   # 凡例を右下に表示 (L)
plt.show()                      # グラフを表示 (M)
```

# 分布の表示

# 行列の数値を等高線で表す

```python
# list B-3-(1)

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# 関数定義 (A)
def myfunc2d(x0, x1):
    y = x0**2 + x1**2 + x0*x1
    return y

# 行列ysの作成 (B)
n0 = 7
n1 = 5
x0s = np.linspace(-3, 3, n0)  # 7個のx0を生成 (C)
x1s = np.linspace(-2, 2, n1)  # 5個のx1を生成 (D)
ys = np.zeros((n1, n0))       # yを格納する行列ys (E)
for i0  in range(n0):
    for i1 in range(n1):
        # 全てのx0とx1の組み合わせに対してyを計算しysに格納 (F)
        y = myfunc2d(x0s[i0], x1s[i1])
        ys[i1, i0] = y

print(ys)
```

```python
# list B-3-(2)

plt.contour(x0s, x1s, ys)
plt.show()
```

```python
# list B-3-(3)

# 行列ysの作成
n0 = 50  # (A)
n1 = 50  # (B)
x0s = np.linspace(-3, 3, n0)
x1s = np.linspace(-2, 2, n1)
ys = np.zeros((n1, n0))
for i0  in range(n0):
    for i1 in range(n1):
        ys[i1, i0] = myfunc2d(x0s[i0], x1s[i1])

# 等高線プロット(C)
cont = plt.contour(
    x0s, x1s, ys, 
    cmap='pink',  # カラーマップ (D)
    levels=[0.2, 0.5, 1, 2, 4, 8, 10],  # 等高線を各レベル (E)
    )
cont.clabel(fmt='%.1f', fontsize=12)  # 等高線に数値を入れる (F)
plt.show()
```
